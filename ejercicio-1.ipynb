```python
import pandas as pd
import seaborn as sb
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Cargar los datos
df = pd.read_csv('housing.csv')

# Eliminar registros con valores nulos
datos = df.dropna()

# Crear dummies para la columna 'ocean_proximity'
dummies = pd.get_dummies(datos['ocean_proximity'], dtype=int)
datos = pd.concat([datos, dummies], axis=1)
datos.drop('ocean_proximity', axis=1, inplace=True)
```

```python
datos['room_ratio'] = datos['total_rooms'] / datos['total_bedrooms']

datos['income_per_latitude'] = datos['median_income'] / datos['latitude']
datos['house_value_per_latitude'] = datos['median_house_value'] / datos['latitude']
```

```python
# Eliminar registros 
datos = datos[datos['median_house_value'] < 500000]
datos = datos[datos['housing_median_age'] < 50]
datos = datos[datos['median_income'] < 15]
```

```python
datos
```

```python

# caracteristicas de entrada
X = datos.drop('median_house_value', axis=1) #axis = 1, para eliminar la columna

#etiqueta
y = datos['median_house_value']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)

modelo = LinearRegression()

#entrenamiento
modelo.fit(X_train, y_train)

predicciones = modelo.predict(X_test)

comparativa = {"predicciones":predicciones, "valor real": y_test}
result = pd.DataFrame(comparativa)

# print(result)

print(f'Score de entrenamiento: {modelo.score(X_train, y_train)}')
print(f'Score de prueba: {modelo.score(X_test, y_test)}')
```
